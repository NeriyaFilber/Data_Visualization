```python
import yfinance as yf
import pandas as pd
```

```python
ticker = ["AAPL", "BA", "KO", "IBM", "DIS", "MSFT",'TSLA','GOOG']
```

```python
stocks = yf.download(ticker,start="2010-01-01",end="2024-06-21")
```

```python
stocks.head(10)
```

```python
stocks.describe()
```

```python
# calculate the cumulative daily returns.
daily_returns = stocks.pct_change()
cumulative_returns = (1 + daily_returns).cumprod()-1
stocks.columns = stocks.columns.get_level_values(1)
cumulative_returns.columns = cumulative_returns.columns.get_level_values(1)
stocks_with_cumulative = stocks.copy()
for i in ticker:
    if ticker in stocks.columns:
        stocks_with_cumulative[f'{i}_Cumulative_Return'] = cumulative_returns[ticker]
    else:
        print(f"Ticker {i} not found in data columns.")
```
