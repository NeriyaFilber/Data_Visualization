```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Vertical Bar Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

plt.barh(categories, values)
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)  # Example data

plt.hist(data, bins=20)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(100, 3)  # Example data

plt.boxplot(data)
plt.xlabel('Variables')
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```

```python
import seaborn as sns
import numpy as np

data = np.random.randn(1000)  # Example data

sns.kdeplot(data)
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('KDE Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.fill_between(x, y, alpha=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Area Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 15, 35]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Plot')
plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import seaborn as sns

# Create a DataFrame with example data
df = pd.DataFrame({'X': np.random.randn(100), 'Y': np.random.randn(100)})

sns.scatterplot(x='X', y='Y', data=df)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('scatter plot')
```

```python
import pandas as pd
import seaborn as sns
import numpy as np

# Create a DataFrame with example data
df = pd.DataFrame({'X': np.random.randn(100), 'Y': np.random.randn(100)})

sns.jointplot(x='X', y='Y', data=df, kind='hex')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Hexbin Plot')
plt.show()
```
